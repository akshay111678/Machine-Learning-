```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
df=pd.read_csv("house.csv")
```

```python
df.head()
```

```python
df['bedrooms'].unique()
```

```python
df['bedrooms']=df['bedrooms'].astype(str)
```

```python
a=df['bedrooms'].unique()
b=df['bedrooms'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(a,b)
plt.title(" Types of BHK")
plt.xticks(rotation=30)
plt.show()
```

```python
df.hist(figsize=(20,20))
plt.show
```

```python
mean=df['price'].mean()
median=df['price'].median()

plt.figure(figsize=(7,7))
plt.hist(df['price'],bins=10)
plt.axvline(mean,color='yellow')
plt.axvline(median,color='yellow')
plt.title("price")
plt.xticks(rotation=30)
plt.show()
```

```python
df['floors']=df['floors'].astype(str)
```

```python
c=df['floors'].unique()
d=df['floors'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(c,d)
plt.title("Flats available on Different Floors")
plt.xticks(rotation=30)
plt.show()
```

```python
df['floors'].value_counts().plot(kind='bar')
```

```python
e=np.full((13,),0.2)
df['bedrooms'].value_counts().plot.pie(autopct='%1.1f%%',explode=e)
```

```python
corr=df.corr()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x=df['sqft_living'],y=df['price'])
plt.title('price vs sqft_living')
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(y=df['bathrooms'],x=df['zipcode'])
plt.title('bathrooms vs zipcode')
plt.ylabel('bathrooms')
plt.xlabel('zipcode')
plt.show()
```
