```python
plt.figure(figsize=(7,7))
plt.scatter(x=df["yr_built"],y=df["price"],color="red")
plt.title("Years built vs Price Scatter plot")
plt.ylabel("price")
plt.xlabel("yr_built")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['bathrooms'],df['price'],color="blue")
plt.scatter(df['bedrooms'],df['price'],color="red")
plt.legend(labels=["bathrooms","bedrooms"])
plt.show()
```

```python
price_mean = df['price'].mean()
```

```python
price_mean
```

```python
plt.figure(figsize=(7,7))
plt.xlim((0,4000000))
plt.hist(df['price'],bins=30)
plt.axvline(price_mean,color="yellow")
plt.axvline(df['price'].median(),color="black")
plt.legend(labels=['Mean','Median'])
plt.show()
```
