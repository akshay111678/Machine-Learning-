### 2) dealing with skewness

```python
df3.describe()
```

```python
from scipy.stats import skew
```

```python
df3['OverallQual']=df3['OverallQual'].astype('str')
df3['OverallCond']=df3['OverallCond'].astype('str')

df3_num=df3.select_dtypes(['int64','float64'])
```

```python
df3_num.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df3_num.set_index('Id',inplace=True)
```

```python
df3_num.head()
```

```python
for col in df3_num:
    plt.figure()
    print(col,skew(df3_num[col]))
    sns.distplot(df3_num[col])
    plt.show()
```

```python
for col in df3_num:
    if skew(df3_num[col])>=1 or skew(df3_num[col])<=1:
        df3_num[col]=np.sqrt(df3_num[col])
```
