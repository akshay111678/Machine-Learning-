# Data Visualization
## matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
df = pd.read_csv("Automobile_data.csv")
```

## LINEPLOT

```python
x=[1,2,3,4]
y=[10,11,12,13]
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(1,10,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.show()
```

```python
area=[600,1000,1200,1800]
price=[15,30,50,60]
plt.plot(area,price)
plt.title("area plot")
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title("area plot")
plt.xlabel('area')
plt.ylabel('price')
plt.xlim(200,2000)
plt.ylim(10,100)
plt.show()
```

```python
x=np.arange(10,50)
y1=np.sqrt(x)
y2=np.log(x)
plt.figure(figsize=(7,7))
plt.plot(x,y1)
plt.plot(x,y2)
plt.title("mathfunction")
plt.xlabel('values')
plt.ylabel('mathfunction')
plt.legend(labels=["sqrt",'log'])
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.hist(df['width'],bins=5)
plt.title("width")
plt.xticks(rotation=30)
plt.show()
```

```python
mean=df['width'].mean()
median=df['width'].median()

plt.figure(figsize=(7,7))
plt.hist(df['width'],bins=10)
plt.axvline(mean,color='yellow')
plt.axvline(median,color='yellow')
plt.title("width")
plt.xticks(rotation=30)
plt.show()
```

```python
mean=df['price'].mean()
median=df['price'].median()

plt.figure(figsize=(7,7))
plt.hist(df['price'],bins=10)
plt.axvline(mean,color='yellow')
plt.axvline(median,color='yellow')
plt.title("price")
plt.xticks(rotation=30)
plt.show()
```

```python
a=df['body-style'].unique()
b=df['body-style'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(a,b)
plt.title("types of car")
plt.xticks(rotation=30)
plt.show()
```

```python
df['body-style'].value_counts().plot(kind='bar')
plt.show()
```

```python
explode=(0.2,0.2,0.20,0.2,0.8)
df['body-style'].value_counts().plot.pie(autopct='%1.1f%%',explode=explode)
```

```python
df['make'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(7,7))
a=np.full((22,),0.4)
df['make'].value_counts().plot.pie(autopct='%1.1f%%', explode=a)
```

# Scatter plot
### it is used to compare to function and find co relation between them

```python
plt.figure(figsize=(7,7))
plt.scatter(df['highway-mpg'],df['price'])
plt.xlabel('highway mpg')
plt.ylabel('price')
plt.title('effect of highway mpg on price')
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['width'],df['price'])
plt.xlabel('width')
plt.ylabel('price')
plt.title('effect of width on price')
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['engine-size'],df['price'])
plt.xlabel('engine size')
plt.ylabel('price')
plt.title('effect of engine size on price')
plt.xticks(rotation=30)
plt.show()
```
