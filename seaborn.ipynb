# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Automobile_data.csv")
```

### Histogram

```python
plt.figure(figsize=(6,6))
sns.distplot(df['width'])
plt.show()
```

```python
mean=df['width'].mean()
median=df['width'].median()
plt.figure(figsize=(6,6))
sns.distplot(df['width'])
plt.axvline(mean,color='yellow')
plt.axvline(median,color='red')
plt.show()
```

### Barplot

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df,y='make')

plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df,x='body-style',hue="fuel-type")

plt.show()
```

### Heatmap

```python
corr=df.corr()
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```

### Scatterplot

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df,x='width',y="price",hue='body-style')
plt.plot()
```

### Boxplot

```python
sns.boxplot(data=df,x='price')
'''the flat line is upper and lower whisker and point which are outside whiskars are called outliars'''
```

```python
sns.boxplot(data=df,x='price',y='make')
```

### Subplot

```python
fig,axes=plt.subplots(2,2,figsize=(9,9))
sns.boxplot(data=df,x='price',ax=axes[0,0])
sns.scatterplot(data=df,x='width',y="price",hue='body-style',ax=axes[0,1])
sns.countplot(data=df,y='make',ax=axes[1,0])
sns.distplot(df['width'],ax=axes[1,1])
plt.show()
```
