# merge

```python
d1 = {"name":["Kalpesh","Kiran","Nimesh","Jitesh"],"math":[77,76,89,87]}
```

```python
d2 = {"name":["Kalpesh","Jitesh","Nimesh","Sushma"],"science":[79,86,82,84]}
```

```python
df11 = pd.DataFrame(d1)
df22 = pd.DataFrame(d2)
df11
```

```python
a=pd.merge(df11,df22)
```

```python
a
```

```python
pd.merge(df11,df22, on="name") #inner
```

```python
pd.merge(df11,df22, on="name", how="outer")
```

```python
a=pd.merge_ordered(df11,df22,fill_method='bfill')
```

```python
a.interpolate()
```
